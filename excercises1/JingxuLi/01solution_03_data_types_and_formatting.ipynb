## Exercise 1: Fix Type Issues

Load the seaborn mpg dataset. Identify and fix all type issues (hint: check horsepower).


```python
import pandas as pd
import numpy as np

df = pd.read_csv('../dataset/mpg.csv')
print("Original dtypes:")
print(df.dtypes)
print(f"\nHorsepower unique samples: {df['horsepower'].sample(10).tolist()}")

df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
print(f"horsepower NA count: {df['horsepower'].isna().sum()}")

print("\nFixed dtypes:")
print(df.dtypes)
```

## Exercise 2: Convert model_year to Datetime

The mpg dataset has a model_year column stored as integers (e.g. 70 = 1970). Convert it to a proper datetime (use January 1 of that year).


```python
df['model_year_dt'] = pd.to_datetime(df['model_year'] + 1900, format='%Y')
print(df[['model_year', 'model_year_dt']].drop_duplicates().head(10))
print(f"\nNew dtype: {df['model_year_dt'].dtype}")
```

## Exercise 3: Extract Manufacturer & Memory Savings

The name column contains strings like 'chevrolet chevelle malibu'. Extract the manufacturer (first word) into a new column and convert it to category dtype. How much memory does this save?


```python
df['manufacturer'] = df['name'].str.split().str[0]
print(f"Unique manufacturers: {df['manufacturer'].nunique()}")
print(df['manufacturer'].value_counts())

obj_memory = df['manufacturer'].memory_usage(deep=True)
df['manufacturer_cat'] = df['manufacturer'].astype('category')
cat_memory = df['manufacturer_cat'].memory_usage(deep=True)

saved_mb = (obj_memory - cat_memory) / 1024 / 1024
print(f"\nMemory (object dtype): {obj_memory / 1024:.2f} KB")
print(f"Memory (category dtype): {cat_memory / 1024:.2f} KB")
print(f"Memory saved: {saved_mb:.4f} MB ({saved_mb/obj_memory*1024*1024*100:.1f}%)")
```

## Exercise 4: String Normalization Function

Write a function normalize_string_column(series) that applies strip, lower, and collapses internal spaces, then apply it to the name column.


```python
def normalize_string_column(series):
    return series.str.strip().str.lower().str.replace(r'\s+', ' ', regex=True)

df['name_normalized'] = normalize_string_column(df['name'])
comparison = pd.DataFrame({'Original': df['name'].head(10).values,
                           'Normalized': df['name_normalized'].head(10).values})
comparison
```
